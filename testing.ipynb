```python
import torch 
import torch.nn as nn
import torch.nn.functional as f
```

```python
device= torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLP,self).__init__()
        self.hidden_layers= nn.ModuleList() #list of connected layers
        in_size= input_size

        for h in hidden_size:
            self.hidden_layers.append(nn.Linear(in_size,h))
            in_size=h
        self.output_layer = nn.Linear(in_size, output_size)
    
    def forward(self,x):
        for layer in self.hidden_layers:
            x= f.relu(layer(x))
        x= self.output_layer(x)
        return x
```

```python
model= MLP(input_size=10, hidden_size=[64,32],output_size=1)

x= torch.randn(5,10)
 
output= model(x)
```
